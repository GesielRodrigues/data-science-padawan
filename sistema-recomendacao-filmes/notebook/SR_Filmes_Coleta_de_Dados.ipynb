```python
def get_id_movie(title_movie):
  result = tmdb_movie.search(title_movie)
  movie_id = result[0].id
  return movie_id
```

```python
def get_features_movie (movie_id, feature):
  
  # Fazendo a requisição
  response_movie = requests.get("https://api.themoviedb.org/3/movie/{}?api_key={}".format(movie_id, tmdb.api_key))
  json_movie = response_movie.json()

  return json_movie[feature]



def get_features_keywords(movie_id, feature):
  
  # Fazendo a requisição
  response_keywords = requests.get("https://api.themoviedb.org/3/movie/{}/keywords?api_key={}".format(movie_id, tmdb.api_key))
  json_keywords = response_keywords.json()

  return json_keywords[feature]



def get_features_credits(movie_id, feature):
  # Fazendo a requisição
  response_credits = requests.get("https://api.themoviedb.org/3/movie/{}/credits?api_key={}". format(movie_id, tmdb.api_key))
  json_credits = response_credits.json()
  return json_credits[feature]
 
```

```python
features_movie = ['budget', 'genres', 'homepage', 'original_language',
            'original_title', 'overview', 'popularity', 'production_companies',
            'production_countries', 'release_date', 'revenue', 'runtime',
            'spoken_languages', 'status', 'tagline', 'title', 'vote_average',
            'vote_count']

features_credits = ['cast', 'crew']

features_keywords = 'keywords'
```

```python
# Corrigindo o titulo do filme The Nan Movie

dfs_concat.loc[dfs_concat['id'] == 647808, 'original_title'] = 'The Nan Movie'	
```

```python
# Coletando as informações de movie da API
for f in features_movie:
  dfs_concat[f] = dfs_concat['id'].map(lambda x : get_features_movie(x, f))
```

```python
# Coletando as informações de credits da API
for f in features_credits:
  dfs_concat[f] = dfs_concat['id'].map(lambda x : get_features_credits(x, f))
    
```

```python
# Coletando as informações de keywords da API
dfs_concat[features_keywords] = dfs_concat['id'].map(lambda x : get_features_keywords(x, features_keywords))
```
