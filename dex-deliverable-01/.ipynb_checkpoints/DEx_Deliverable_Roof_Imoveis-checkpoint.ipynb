## 4. A Limpeza de Dados

### 4.1 Adicionando algumas colunas que complementam as informa√ß√µes e ajustando a coluna view

Ajustando o formato das datas para o tipo datetime que √© mais f√°cil de utilizar.

```python
df['date'] = pd.to_datetime(df['date'])
df.head()
```

Calcular o pre√ßo por √°rea de im√≥vel e incluir em uma nova coluna chamada sqrt_price, pra isso vamos somar o sqft_living com sqft_lot e dividir pelo pre√ßo.

```python
# Criando a coluna com o pre√ßo por √°rea:
df['sqft_total'] = df['sqft_living'] + df['sqft_lot']
df['sqft_price'] = df['price'] / df['sqft_total']

df.head()
```

```python
# Baixando o Dataset de cidades por zipcodes do Google Drive, dataset criado a partir do site: https://www.zip-codes.com/county/wa-king.asp

!gdown --id 14TNDm5DrB-tRlvby0Eynjny96cZkneb8
```

```python
# Lendo o dataset de cidades por zipcode:
zipcodes = pd.read_csv('kc_zipcodes.csv', delimiter = ';')

# Transformando o dataset em um dict para usarmos na fun√ß√£o map:
zipcodes = zipcodes.set_index('ZIP Code').T.to_dict('list')
print(zipcodes)

# Tratamento do dict, retirando os colchetes e aspas para ficar apenas o nome da cidade na coluna do dataset:
for k in zipcodes.keys():
  zipcodes[k] = zipcodes.get(k)[0]

# Criando a coluna city com o nome das cidades:
df['city'] = df['zipcode'].map(zipcodes)

df.head()
```

```python
# Criando uma coluna com a idade das casas, fazendo com o ano de constru√ß√£o e o ano de venda:
df['age'] = df['date'].dt.year - df['yr_built']

df.head()
```

```python
# Classificar a idade da reforma em categorias (recente, 2 a 4 anos atr√°s, 5 a 9 anos, 10 a 24 anos, mais de 25 anos e nunca reformada)

df['renovation_age'] = df.apply(lambda x : x.date.year - x.yr_renovated if x.yr_renovated > 0 else 'Never', axis=1)

def renovation_status(obs):
    status = ''
    x = obs['renovation_age']
    if(x == 'Never'):
        status = 'Never'
    elif(x > 24):
        status = '25+'
    elif(x > 9):
        status = '10-24'
    elif(x > 4):
        status = '5-9'
    elif(x > 1):
        status = '2-4'
    elif(x == 1 or x == 0 or x == -1):
        status ='Recent'
    return status

df['renovation_age'] = df.apply(renovation_status, axis=1)
df.renovation_age.value_counts()
```

### 5.2 Montando um datase apenas com as casas que foram negociadas mais de uma vez

```python
df = df.sort_values(['id', 'date'])
# Dividir em 2 dataset o primeiro com a primeira negocia√ß√£o da casa e o segundo com a √∫ltima
df_aux = df[df.id.duplicated(keep = 'last')]
df_aux2 = df[df.id.duplicated(keep = 'first')]

df_aux = df_aux.drop_duplicates('id', keep='first')
df_aux2 = df_aux2.drop_duplicates('id', keep='last')

# Deixando apenas as colunas de id, date e price no segundo dataset
df_aux2 = df_aux2.drop(['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'viewed', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15', 'sqft_total',	'sqft_price',
       'city', 'age',	'renovation_age'], axis=1)

# Excluindo linhas duplicadas da casa que teve 3 negocia√ß√µes, ficando apenas a primeira e √∫ltima
df_aux = df_aux.sort_values('id')
df_aux2 = df_aux2.sort_values('id')


# Unindo os datasets em um, com a coluna date e price da √∫ltima negocia√ß√£o nas √∫ltimas colunas
df_negociadas = df_aux.join(df_aux2.set_index('id'), on = 'id', rsuffix = '_2nd')


# Reorganizar as colunas
columnsTitles = ['id', 'date', 'price', 'date_2nd', 'price_2nd', 
                 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
                 'floors', 'waterfront', 'viewed', 'condition', 'grade',
                 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated',
                 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15', 'sqft_total',	'sqft_price',
                 'city', 'age',	'renovation_age']

# Renomeando as colunas
df_negociadas = df_negociadas.reindex(columns=columnsTitles)
df_negociadas.rename(columns = {'date_x': '1st_date', 
                                'price_x' : '1st_price', 
                                'date_y' : '2nd_date', 
                                'price_y' : '2nd_price'}, inplace = True)

df_negociadas.head()
```

Primeiro vamos calcular o retorno do investimento nas casas negociadas mais de uma vez:

Rentabilidade = Rendimento L√≠quido x 100 / Valor Investido

Depois iremos dividir pelo n√∫mero de dias, para um retorno por dia, e incluir esse retorno em uma coluna.



```python
# Calculando a quantidade de dias de uma venda para a outra
days = (df_negociadas['date_2nd'] - df_negociadas['date']).astype('timedelta64[D]').astype(int)

# Calculando o retorno do neg√≥cio
df_negociadas['rent_total'] = ((df_negociadas['price_2nd'] - df_negociadas['price']) / df_negociadas['price']) * 100

# Divis√£o do retorno total pelos dias entre as vendas para o retorno di√°rio
df_negociadas['rent_daily'] = df_negociadas['rent_total'] / days


df_negociadas.head()
```

## 6. An√°lise final e sugest√£o de im√≥veis

As casas mais caras est√£o localizadas em Medina, mais especificamente no zipcode 98039.

Medina tamb√©m √© o local que tem o maior pre√ßo por √°rea e em seguida vem Seattle.

Seattle tem o menor tamanho m√©dio de casas.

Casas com vista para o mar s√£o praticamente 3x mais valorizadas dos que as casas sem vista.

Como era de se esperar, alem desses fatores acima (localiza√ß√£o e vista para o mar) os pre√ßos das casas est√£o altamente correlacionados com a √°rea habit√°vel (sqft_living), com a condi√ß√£o (condition) e tambem com a qualidade dos materiais utilizados na casa (grade).

Quando analisamos apenas as casas que tiveram mais de uma negocia√ß√£o, para avaliarmos o retorno que o investimento teve, encontramos algo que parece contra intuit√≠vo mas, que pensando bem, faz bastante sentido: As casas mais baratas tiveram maior retorno e as mais caras tiveram um retorno menor (e at√© negativo).

Ou seja, na hora do investimento, um ponto importante foi pagar pouco na casa para revender com lucro posteriormente.

Dessa forma faremos um ranking por casas mais baratas, que estejam em boa condi√ß√£o (condition > 3) e que tenham sido utilizados materiais de qualidade (grade acima de 5). Al√©m disso vamos dar um peso maior para casas localizadas em Seattle.

Desse ranking n√≥s tiraremos as primeiras 5 as casas que s√£o as que devemos comprar primeiro.
E as 5 √∫ltimas seriam aquelas que n√£o devemos comprar de forma alguma.

```python
# Vamos excluir as colunas que n√£o far√£o parte da nossa an√°lise:

# df.columns

colunas_para_exclus√£o = ['date', 'bedrooms', 'bathrooms', 'sqft_living',
                         'sqft_lot', 'floors', 'waterfront', 'viewed',
                         'sqft_above', 'sqft_basement', 'yr_built',
                         'yr_renovated', 'zipcode','lat', 'long',
                         'sqft_living15', 'sqft_lot15', 'sqft_total',
                         'sqft_price', 'age', 'renovation_age']

df = df.drop(colunas_para_exclus√£o, axis=1)

df.head(10)
```

```python
# Ordenando o dataset pelo pre√ßo
df.sort_values(by = 'price', inplace = True)

# Fazendo o ranqueamento
df['price_ranking'] = range(1, df.shape[0] + 1)

df
```

```python
# Ranquear a condi√ß√£o, quanto maior a condi√ß√£o menor o ranking (melhor a posi√ß√£o)
# Condition vai de 1 a 5 e receber√° notas de 10 a 50 em ordem decrescente


def ranking_condition(x):
    condition = 50
    if x == 5:
        condition = 10
    elif x == 4:
        condition = 20
    elif x == 3:
        condition = 30
    elif x == 2:
        condition = 40
    else:
        condition = 50
    
    return condition



df['condition_ranking'] = df['condition'].apply(ranking_condition)
df.condition_ranking.value_counts()
```

```python
# Ranquear a grade, quanto maior a grade menor o ranking (melhor a posi√ß√£o)
# Condition vai de 1 a 13 e receber√° notas de 10 a 30 sendo 10 grade acima de 10, 20 grade entre 5 e 9 e 30 grades abaixo de 5

def ranking_grade(g):
    grade = 30
    if g >= 10:
        grade = 10
    elif g >= 5:
        grade = 20
    else:
        grade = 30
    
    return grade



df['grade_ranking'] = df['grade'].apply(ranking_grade)
df.grade_ranking.value_counts()
```

```python
# O ranking de cidades vamos dar nota 10 para Seattle, 20 para Federal Way, Kent ou Renton e as demais peso 30

def ranking_city(c):
    rank = 30
    if c == 'Seattle':
        rank = 10
    elif c == 'Federal Way' or c == 'Kent' or c == 'Renton':
        rank = 20
    else:
        rank = 30
    
    return rank



df['city_ranking'] = df['city'].apply(ranking_city)
df.city_ranking.value_counts()
```

```python
# Agora faremos o ranking total somando todos os rankings

df['ranking'] = df['price_ranking'] + df['condition_ranking'] + df['grade_ranking'] + df['city_ranking']

df
```

```python
# Ordenando do menor ranking para o maior
df = df.sort_values(by = 'ranking')
```
